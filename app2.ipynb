pyファイル2つおよびckptファイルをアップロードしている状態で以下のセルを実行するとngrokが実行されリンクが生成されます。このとき、3行目にあるトークンはngrokに登録することで取得できます。
以下、トークンの取得手順を記載します。


1.   https://ngrok.com/ にアクセスする
2.   sign upより新規登録を行う(githubまたはGoogleのアカウントで可能)
3.   登録完了後、左上にある"Your Authtoken"をクリックし、画面上部の伏字になっている文字列を下のコード内に入力する







```python
from pyngrok import ngrok,conf

token = "トークンを入力してください"
!pkill -f streamlit
!streamlit run app.py --server.port 8501 &>/dev/null &
!pgrep streamlit

from pyngrok import ngrok

!pkill ngrok

ngrok.set_auth_token(token)

#Streamlitが使用するデフォルトのポート番号8501を明示的に指定
public_url = ngrok.connect("8501").public_url
def convert_http_to_https(url):
    # urlの先頭がhttpだったらhttpsに変更
    if url.startswith("http://"):
        # 1は置換する回数(1を指定することで、元のURL内で最初に見つかった "http://" の部分だけを "https://" に置き換え、他の "http://" は変更されない。)
        url = url.replace("http://", "https://", 1)
    return url
print(convert_http_to_https(public_url))
```
